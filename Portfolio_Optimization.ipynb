```python
import yfinance as yf
import datetime as dt
import numpy as np
import pandas as pd
#將不同的個股收盤價裝入dataframe中
Closeprice = pd.DataFrame()
#決定投資標的
tickers = ['2801.TW','2880.TW','2881.TW','2882.TW','2884.TW','2886.TW','2891.TW','2892.TW','5880.TW']
#迴圈下載個股收盤價
for ticker in tickers:
  tmp = yf.download(ticker, start='2020-01-01', end='2021-06-20')
  Closeprice[ticker] = tmp['Close']
  returns = Closeprice
```
